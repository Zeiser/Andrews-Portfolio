```python
activistCo = list(sw50['Name'])
```

```python
activistCo
```

```python
splitNameCO= [x.split(' ')[:-1] for x in activistCo]
```

```python
splitNameCO
```

```python
CoName = []

for x in splitNameCO:
    if not (x[-1].endswith(')') or x[-1].startswith('(') or x[-1].endswith('.')):
        CoName.append(x)
    else:
        CoName.append(x[:-1])

splitNameCO = CoName   

splitNameCO
```

```python
CoName = []

for x in splitNameCO:
    if x[-1].endswith(','):
        x[-1] = x[-1][:-1]
    CoName.append(x)
splitNameCO = CoName   

splitNameCO    
```

```python

JoinNameCO = [' '.join(x) for x in splitNameCO]
```

```python
JoinNameCO
```

```python
#creating word bank to spot sharkwatch50 leaders
words ='|'.join(activistSurnames + JoinNameCO)
```

```python
# identifiy all sharks from the word bank and using it to identify if they are filing new forms
```

```python
activist_filings = idx.working_idx
```

```python
for index , row in activist_filings.iterrows():
    print(row.Filename)
    g = sec.filingArchive(row.Filename, datadir = '.\secdata', user_agent = user_agent_str)
    g.get_filingArchive()
    file = g.files[0]
    matches= re.findall(words,file.body,re.IGNORECASE)
    print(matches)
    if any(matches):
        names = [x for x in set(matches)]
        #names = [x ]
        print(names)
        activist_filings.at[index,'Activist'] = names
```
