```python
#importing Trade and Quotes data
import pandas as pd
TRADES_FILE = 'EQY_DJIA_TRADE_20190814.p'
QUOTES_FILE = 'EQY_DJIA_QUOTES_20190814.p'
dtrades = pd.read_pickle(TRADES_FILE)
dquotes = pd.read_pickle(QUOTES_FILE)
```

```python
#making the data visable between a time frame
tradeTime = dtrades.between_time('9:30', '16:00')
quoteTime = dquotes.between_time('9:30', '16:00')
```

```python
sQ = tradeTime[tradeTime.Symbol == 'AAPL']
```

```python
sT = quoteTime[quoteTime.Symbol == 'AAPL']
```

```python
# merging the trades and quotes data set
aapl = pd.merge_asof(sT, sQ, left_index = True, right_index = True)
```

```python

aapl.head()
```

```python
# calculating financal calculations to better understand stocks
aapl['midpoint']= (aapl['Best_Bid_Price'] + aapl['Best_Offer_Price'])/2

aapl.head()
```

```python
aapl['Percent Quoted Spread']=  ((aapl['Best_Offer_Price'] - aapl['Best_Bid_Price']) / aapl['midpoint']) 
```

```python
aapl.head()
```

```python
aapl['Percent Effective Spread']=  (2 * abs(aapl['Trade Price'] - aapl['midpoint'])) / aapl['midpoint'] 
```

```python
aapl.head()
```

```python
aapl['Depth'] = aapl['Best_Bid_Size'] + aapl['Best_Offer_Size']
```

```python
aapl.head()
```

```python
def pit(pes,pqs):
    if pes < pqs:
        return 1
    else:
        return 0 
    
```

```python
aapl['Price Improvement'] = aapl.apply(lambda x: pit(x['Percent Effective Spread'],x['Percent Quoted Spread']), axis = 1)
```

```python
aapl.head()
```

```python
def TT(pes,pqs):
    if pes > pqs:
        return 1
    else:
        return 0 
```

```python
aapl['Trade Through'] = aapl.apply(lambda x: TT(x['Percent Effective Spread'],x['Percent Quoted Spread']), axis = 1)
```

```python
#creating a function that that builds a summarizing dataframe which can used multiple times to answer questions before investing  


groupedQuotes = quoteTime.groupby('Symbol')
groupedTrades = tradeTime.groupby('Symbol')
def sumBuild(tick):
    groupedTrades = tradeTime.groupby('Symbol')
    dt = groupedTrades.get_group(tick)
    dq = groupedQuotes.get_group(tick)
    tradeQuote = pd.merge_asof(dt, dq, left_index = True, right_index = True)
    tradeQuote.dropna()
    tradeQuote['midpoint']= (tradeQuote['Best_Bid_Price'] + tradeQuote['Best_Offer_Price'])/2
    tradeQuote['Percent Quoted Spread']=  ((tradeQuote['Best_Offer_Price'] - tradeQuote['Best_Bid_Price']) / tradeQuote['midpoint']) 
    tradeQuote['Percent Effective Spread']=  (2 * abs(tradeQuote['Trade Price'] - tradeQuote['midpoint'])) / tradeQuote['midpoint']
    tradeQuote['Depth'] = tradeQuote['Best_Bid_Size'] + tradeQuote['Best_Offer_Size']
    tradeQuote['Price Improvement'] = tradeQuote.apply(lambda x: pit(x['Percent Effective Spread'],x['Percent Quoted Spread']), axis = 1)
    tradeQuote['Trade Through'] = tradeQuote.apply(lambda x: TT(x['Percent Effective Spread'],x['Percent Quoted Spread']), axis = 1)
    pqs = tradeQuote['Percent Quoted Spread'].mean()
    pes = tradeQuote['Percent Effective Spread'].mean()
    depth = tradeQuote['Depth'].mean()
    pi = tradeQuote['Price Improvement'].mean()
    tt = tradeQuote['Trade Through'].mean()
    return pqs, pes, depth, pi, tt
```

```python
pqs, pes, depth, pi, tt = sumBuild('AAPL')
```

```python
#developing the dataframe
summary = pd.DataFrame()
groupedQuotes = quoteTime.groupby('Symbol')
groupedTrades = tradeTime.groupby('Symbol')
ticks=groupedQuotes.groups.keys()
for tick in ticks: 
    pqs, pes, depth, pi, tt = sumBuild(tick)
    summary.at[tick, 'Percent Quoted Spread'] = pqs
    summary.at[tick, 'Percent Effective Spread'] = pes
    summary.at[tick, 'Depth'] = depth
    summary.at[tick, 'Trade Throughs'] = tt
    summary.at[tick, 'Price Improvement'] = pi
```
