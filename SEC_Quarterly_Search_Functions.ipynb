```python
## identifiy how many fillings of each form there are. 

#step 1 def a function that parses the lines
    # strip off the \n and turn it into a list
    #turn each colum in the lest into a variable 

def parse(docline):
    docline=docline.rstrip('\n')
    docline=docline.split("|")
    CIK = docline[0]
    companyName = docline[1]
    formType = docline[2]
    dateFiled = docline[3]
    fileName = docline[4]
    return CIK , companyName, formType, dateFiled, fileName
```
